```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
import operator
```

```python
np.random.seed(0)
x_data = np.linspace(-2, 2, num=25)
y_data = 2 * np.sin(1.5 * x_data) + np.random.normal(size=25)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data)
```

```python
x_data = x_data[:, np.newaxis]
y_data = y_data[:, np.newaxis]
```

```python
model = LinearRegression()
model.fit(x_data, y_data)
y_pred = model.predict(x_data)
rmse = np.sqrt(mean_squared_error(y_data,y_pred))
print(rmse)
plt.scatter(x_data, y_data, s=10)
plt.plot(x_data, y_pred, color='r')
plt.show()
```

```python
polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x_data)
model = LinearRegression()
model.fit(x_poly, y_data)
y_poly_pred = model.predict(x_poly)
rmse = np.sqrt(mean_squared_error(y_data,y_poly_pred))
print(rmse)
plt.scatter(x_data, y_data, s=10)
plt.plot(x_data, y_poly_pred, color='m')
plt.show()
```
